```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

**Check for missing values**

```python
#your code here
spaceship.isnull().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
#your code here
spaceship_cleaned = spaceship.dropna()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
#your code here
spaceship_cleaned['Cabin'] = spaceship_cleaned['Cabin'].str[0]
spaceship_cleaned
```

- Drop PassengerId and Name

```python
#your code here
spaceship_cleaned = spaceship_cleaned.drop(columns=['Name', 'PassengerId'])
```

- For non-numerical columns, do dummies.

```python
#your code here
dummies = pd.get_dummies(spaceship_cleaned, columns=['Cabin', 'HomePlanet', 'Destination', 'VIP', 'CryoSleep'], drop_first=True)
dummies.head()
```

**Perform Train Test Split**

```python
# #your code here
features = dummies.drop(columns=['Transported'])
target = dummies['Transported']
X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
#your code here
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

- Evaluate your model's performance. Comment it

```python
#your code here
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score, accuracy_score

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")

rmse = root_mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f"R2: {r2}")

# Insights:
print("""\nThe KNN model shows moderate performance based on the evaluation metrics. The RMSE and MAE values indicate the average prediction error, while the R2 score suggests that the model explains a reasonable portion of the variance in the target variable. 
Further tuning and feature engineering may improve the model's accuracy.
Based on the current metrics, the model performs adequately but has room for improvement.
""")
```
